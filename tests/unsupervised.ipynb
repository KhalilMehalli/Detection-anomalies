```python
import matplotlib.pyplot as plt
import os
import time
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.metrics import roc_curve, auc
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_kddcup99
from sklearn.utils import Bunch

```

```python
def gen_blobs_with_outliers(n_samples, n_outliers_fractions, centers, cluster_std, low, high, random_state=3):
    """
    This function will generate a 2d dataset of :
      - n_samples points sampled with make_blobs (Gaussian distribution)
      - n_outliers points sampled from a uniform distribution over [low, high]
    """
    
    print(n_samples,n_outliers_fractions)
    n_outliers = int(n_samples * n_outliers_fractions)
    print(n_outliers)

    
    # create a blobs of n points 
    X, _ = make_blobs(n_samples, centers=centers,cluster_std=cluster_std, random_state=random_state)

    # Generater to have always the same distribution with the same parameter 
    rng = np.random.RandomState(random_state+n_samples)
    
    outliers = rng.uniform(low=low, high=high, size=(n_outliers, 2))
    X_full = np.vstack([X, outliers])

    # Create label for each x
    y = np.hstack([np.zeros(n_samples, dtype=int), np.ones(n_outliers, dtype=int)])
    return X_full, y

def plot_point(x,y,title):
    mask_outlier = (y==1)
    mask_inlier = ~mask_outlier
    plt.scatter(x[mask_inlier, 0], x[mask_inlier, 1], color="green", label="Inliers")
    plt.scatter(x[mask_outlier, 0], x[mask_outlier, 1], color="red", label="Outliers")
    plt.legend(bbox_to_anchor=(1.02, 1)) # Legend out of the plot if it will hide points
    plt.title(title)

def plot_roc(y_ground_truth, scores,title="ROC Curve"):
    fpr, tpr, _ = roc_curve(y_ground_truth, scores)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f"{title} (AUC = {roc_auc:.3f})")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.legend(loc="lower right")
    plt.title(title)


    
```

```python
# GROUND TRUTH VALUE 

n_samples,n_outliers_fractions,clusters = 1000, 0.15, 2 

x1, y1 = gen_blobs_with_outliers(n_samples,n_outliers_fractions,clusters,0.5,-10,10)
plot_point(x1,y1,f" {n_samples} points in {clusters} blobs and with {n_outliers_fractions*100}% outliers generate with an uniform distribution")
```

### LOF

```python
def lof_test(x,n_neighbors,contamination):
    lof = LocalOutlierFactor(n_neighbors= n_neighbors,contamination= contamination)
    y_pred = lof.fit_predict(x)
    y= (y_pred == -1) # y_pred is -1 or 1, we want a binary y for the plot function 

    scores = -lof.negative_outlier_factor_

    return y, scores

        
def plot_lof_test(x,y_ground_truth, n_neighbors_list,contamination=0.1):
    lof_results = []
    plot_point(x,y_ground_truth, "Ground truth")
    
    for i, n_nb in enumerate(n_neighbors_list):
        plt.figure(figsize=(15, 4)) # Create a new graph each time 

        plt.subplot(1, 2, 1) # This figure will be at the left
        y_pred, scores = lof_test(x,n_nb,contamination)
        lof_results.append(y_pred)
        plot_point(x,lof_results[i],f"LOF with {n_nb} n_neighbors and contamination {contamination}")

        plt.subplot(1, 2, 2) # This figure will be at the right
        plot_roc(y_ground_truth,scores)        
    return lof_results
```

```python
lof_test_results = plot_lof_test(x1,y1,[5,20,50,100,150],n_outliers_fractions)
```

### IForest 

```python
def IForest_test(X, n_tree, contamination):

    start = time.perf_counter()
    IF = IsolationForest(n_estimators=n_tree,contamination=contamination).fit(X) 
    y_pred = (IF.predict(X) == -1)
    elapsed = time.perf_counter() - start
    print(f"Execution time : {elapsed:.4f} s")

    scores = -IF.decision_function(X)
    return y_pred, scores

def plot_IF_test(x,y_ground_truth, n_tree_list,contamination=0.1):
    print(f"Points : {len(x)}")
    
    IF_results = []
    plt.figure()
    plot_point(x,y_ground_truth, "Ground truth")
    plt.show()
    
    for i, n_t in enumerate(n_tree_list):
        plt.figure(figsize=(15, 4)) # Create a new graph each time 

        plt.subplot(1, 2, 1) # This figure will be at the left
        y_pred, scores = IForest_test(x,n_t,contamination)
        IF_results.append(y_pred)      
        plot_point(x,IF_results[i],f"IF with {n_t} trees and contamination {contamination}")

        plt.subplot(1, 2, 2) # This figure will be at the right
        plot_roc(y_ground_truth,scores)
        plt.show()
        
    return IF_results
```

```python
IF_test_results = plot_IF_test(x1,y1,[10,50,100,150,200],n_outliers_fractions)
```

```python
def csv_data_infos(filenames, outlier_label=1):
    for name in filenames:
        data = pd.read_csv(name).to_numpy()
        
        x = data[:,:-1]
        y = data[:,-1]
        
        samples = x.shape[0]
        features = x.shape[1]

        # Transform label for outlier into 1 and 0 for the normal value
        y_bin = (y == outlier_label).astype(int)

        anomaly = np.sum(y_bin)
        p_anomaly = anomaly/ samples

        print({
            'filename' : name,
            'samples'  : samples,
            'features' : features,
            'anomalies': anomaly,
            'p_anomaly': p_anomaly
        })
csv_data_infos(filenames)
```

```python
def barplot(scores,y_ground_truth, color_outlier='red', color_inlier='green', title="Outliers scores " ):
    mask_outlier = (y_ground_truth==1)
    mask_inlier = ~mask_outlier

    # plt.bar need an indices tab
    indices = np.arange(scores.size)

    start = time.perf_counter()
    plt.bar(indices[mask_outlier], scores[mask_outlier], color=color_outlier, edgecolor=color_outlier)
    plt.bar(indices[mask_inlier], scores[mask_inlier], color=color_inlier, edgecolor=color_inlier)
    elapsed = time.perf_counter() - start
    print(f"Plot execution time : {elapsed:.4f} s")
    

    plt.xlabel("index points")
    plt.ylabel("Outliers scores")
    plt.title(title)

y_pred, scores = IForest_test(datasets[0].data, 100, 0.01)
print("Prediction by IFOREST : ", y_pred[:50])
print("Ground Truth of y : ", datasets[0].target[:50])
# TEST WITH ONLY 50 POINTS
barplot(scores[:50],datasets[0].target[:50], title='Outliers scores, 50 firsts points')

plt.figure()
barplot(scores,datasets[0].target)
```
